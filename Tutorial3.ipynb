```python
# Cell 1: imports and shape variables
import numpy as np

# You can change n, p, m to any integers
n = 2
p = 3
m = 4
```

```python
# 1) 2 x 2 array (example)
arr_2x2 = np.array([[1, 2],
                    [3, 4]])
print("2x2 array:\n", arr_2x2)
```

```python
# 2) random 3 x 4 array (values in [0,1))
rand_3x4 = np.random.random((3, 4))
print("Random 3x4 array:\n", rand_3x4)
```

```python
# 3) ones and zeros
ones_3x2 = np.ones((3, 2))
zeros_2x5 = np.zeros((2, 5))
print("Ones 3x2:\n", ones_3x2)
print("Zeros 2x5:\n", zeros_2x5)
```

```python
# 4) random 6x3 and transpose -> becomes 3x6
rand_6x3 = np.random.random((6, 3))
rand_6x3_T = rand_6x3.T
print("Random 6x3:\n", rand_6x3)
print("Transposed (3x6):\n", rand_6x3_T)
```

```python
# 5) dot product of (n x p) and (p x m) -> result (n x m)
A = np.random.random((n, p))
B = np.random.random((p, m))
C = A.dot(B)   # or np.dot(A, B)
print(f"A shape: {A.shape}, B shape: {B.shape}, Result C shape: {C.shape}")
print("A:\n", A)
print("B:\n", B)
print("A dot B = C:\n", C)
```

```python
# integer random example
A_int = np.random.randint(0, 10, (n, p))
B_int = np.random.randint(0, 10, (p, m))
C_int = A_int @ B_int
print("Integer A:\n", A_int)
print("Integer B:\n", B_int)
print("Integer A @ B:\n", C_int)
```
